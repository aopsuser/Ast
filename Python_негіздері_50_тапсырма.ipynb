# Практикалық тапсырмалар: Python негіздері




## 1.Негізгі функциялар

| Функция / Кілт сөз | Сипаттамасы (Қазақша) | Мысал (Python) |
|---|---|---|
| `print()` | Экранға мәтін, айнымалы немесе нәтиже шығарады | `print("Hello, World!")` |
| `type()` | Мәннің немесе айнымалының деректер типін анықтайды | `print(type(x))` |
| `#` | Бір жолдық комментарий жазу үшін қолданылады (оқылмайды) | `# Бұл түсіндірме мәтін` |
| `len()` | Мәтіннің немесе тізімнің ұзындығын санайды | `len("Python") # Нәтиже: 6` |

## 2. Деректер типтері және Түрлендіру (Data Types & Casting)

| Функция | Сипаттамасы (Қазақша) | Мысал (Python) |
|---|---|---|
| `int()` | Бөлшек санды немесе мәтінді бүтін санға айналдырады | `int(3.14) # Нәтиже: 3` |
| `float()` | Бүтін санды немесе мәтінді бөлшек санға айналдырады | `float(5) # Нәтиже: 5.0` |
| `str()` | Кез келген деректер типін мәтінге (жолға) айналдырады | `str(2024) # Нәтиже: "2024"` |
| `bool()` | Мәнді логикалық типке (True/False) айналдырады | `bool(1) # Нәтиже: True` |

## 3. Мәтіндермен жұмыс істеу әдістері (String Methods)


| Әдіс / Синтаксис | Сипаттамасы (Қазақша) | Мысал (Python) |
|---|---|---|
| `[ : ]` | Мәтіннің белгілі бір бөлігін кесіп алу (Slicing) | `"Apple"[0:3] # Нәтиже: "App"` |
| `.upper()` | Мәтіндегі барлық әріптерді бас әріпке айналдырады | `"hi".upper() # Нәтиже: "HI"` |
| `.lower()` | Мәтіндегі барлық әріптерді кіші әріпке айналдырады | `"HI".lower() # Нәтиже: "hi"` |
| `.strip()` | Мәтіннің басы мен соңындағы бос орындарды алып тастайды | `" a ".strip() # Нәтиже: "a"` |
| `.replace(a, b)` | Мәтіндегі бір таңбаны не сөзді екіншісіне ауыстырады | `"Cat".replace("C", "B")` |
| `.split()` | Мәтінді көрсетілген белгі бойынша бөліп, тізім жасайды | `"a,b".split(",") # ['a', 'b']` |
| `.count()` | Мәтін ішінде белгілі бір әріптің/сөздің неше рет кездесетінін санайды | `"banana".count("a") # 3` |
| `f"..."` | Айнымалыларды мәтін ішіне тікелей енгізуге мүмкіндік береді | `f"My age is {age}"` |

## 4. Математикалық және Логикалық операторлар (Operators & Booleans)

| Оператор | Сипаттамасы (Қазақша) | Мысал (Python) |
|---|---|---|
| `+`, `-`, `*`, `/` | Қосу, алу, көбейту, бөлу | `10 / 2 # Нәтиже: 5.0` |
| `**` | Дәрежеге шығару | `3 ** 2 # Нәтиже: 9` |
| `//` | Бүтін бөлу (бөлшек бөлігін алып тастайды) | `10 // 3 # Нәтиже: 3` |
| `%` | Бөлуден қалған қалдықты табу (жұп/тақ тексеруге ыңғайлы) | `10 % 3 # Нәтиже: 1` |
| `==`, `!=` | Тең бе? / Тең емес пе? | `5 == 5 # True` |
| `>`, `<`, `>=`, `<=` | Үлкен, кіші, үлкен немесе тең, кіші немесе тең | `10 >= 8 # True` |
| `and` | Екі шарт та Ақиқат (True) болса ғана орындалады | `x > 0 and x < 10` |
| `or` | Шарттардың біреуі Ақиқат (True) болса жеткілікті | `x == 5 or x == 10` |
| `not` | Логикалық мәнді керісінше өзгертеді (True-ды False-қа) | `not True # Нәтиже: False` |

## 5. Шартты операторлар (Conditions)

| Кілт сөз | Сипаттамасы (Қазақша) | Құрылымы (Мысал) |
|---|---|---|
| `if` | Егер берілген шарт дұрыс (True) болса, төмендегі код орындалады. | `if x > 0: print("Оң сан")` |
| `elif` | (Else if). Алдыңғы if шарты қате болса, келесі жаңа шартты тексереді. | `elif x == 0: print("Нөл")` |
| `else` | Жоғарыдағы шарттардың ешқайсысы орындалмаған жағдайда жұмыс істейді. | `else: print("Теріс сан")` |

```
# 36-тапсырма: f-string көмегімен айнымалыларды мәтінге қосыңыз: name = "Ali", age = 20. Нәтиже: "Ali is 20 years old"
name = "Ali"
age = 20
print(f"{name} is {age} years old")
```

```
# 37-тапсырма: "Алматы" сөзін керісінше жазып шығарыңыз. (Кеңес: [::-1])
print('Алматы'[::-1])
```

```
# 38-тапсырма: "Programming" сөзінде "gram" сөзі бар екенін тексеріңіз (`in` кілт сөзі).
print('gram' in 'Programming')
```

```
# 39-тапсырма: "Python is awesome" сөзін бос орындар арқылы тізімге бөліңіз (`split()`).
print('Python is awesome'.split())
```

```
# 40-тапсырма: `str()` функциясын пайдаланып 2024 санын мәтінге айналдырып, оған " жыл" сөзін қосып шығарыңыз.
num = 2024
nenum = str(num)
print(nenum + " жыл")
```



## 4-бөлім: Логика және Шартты операторлар (If, Elif, Else) (41-50 тапсырмалар)

Бағдарламаның логикасын құрып, белгілі бір шарттарға байланысты әртүрлі шешімдер қабылдаймыз.


```
# 41-тапсырма: 10 мен 5-ті салыстырыңыз (10 > 5). Нәтижесін шығарыңыз.
print(10 > 5)
```

```
# 42-тапсырма: 15 пен 15-тің тең екенін тексеріңіз (`==`).
print(15 == 15)
```

```
# 43-тапсырма: `x = 10`. Егер `x > 5` болса, "Үлкен" деп шығаратын шарт жазыңыз.
x = 10
if x > 5:
  print("Үлкен")
```

```
# 44-тапсырма: `num = 4`. `if-else` арқылы бұл санның жұп немесе тақ екенін тексеріп шығарыңыз (қалдық `% 2 == 0` болуы керек).
num = 4
if num % 2 == 0:
  print('жұп')
else:
  print('тақ')
```

```
# 45-тапсырма: `a = 15`, `b = 20`. Қайсысы үлкен екенін `if...else` арқылы табыңыз.
a, b = 15, 20
if a>b:
  print(str(a), "үлкен")
else:
  print(str(b), "үлкен")
```

```
# 46-тапсырма: `score = 75`. Егер баға >= 90 болса "A", >= 80 болса "B", >= 70 болса "C", әйтпесе "D" деп шығаратын код жазыңыз (`if...elif...else`).
score = 75
if score >= 90:
  print("A")
elif score >= 80:
  print("B")
elif score >= 70:
  print("C")
else:
  print("D")
```

```
# 47-тапсырма: `password = "secret123"`. Егер password дәл осы мәнге тең болса "Қосылды", әйтпесе "Қате құпия сөз" деп шығарыңыз.
password = "secret123"
passw = input()
if passw == password:
  print("Қосылды")
else:
  print("Қате құпия сөз")
```

```
# 48-тапсырма: `age = 16`. 18-ге толғанын тексеріңіз. Егер толмаса, кәмелетке толуға қанша жыл қалғанын есептеп шығарыңыз.
age = 16
if age >= 18:
  print("Malades")
else:
  print(18 - age)
```

```
# 49-тапсырма: Логикалық `and` қолдану: `x = 7`. Егер x > 0 ЖӘНЕ x < 10 болса "Бір таңбалы оң сан" деп шығарыңыз.
x = 7
if x > 0 and x < 10:
  print("Бір таңбалы оң сан")
```

```
# 50-тапсырма: Логикалық `or` қолдану: `day = "Сенбі"`. Егер day "Сенбі" НЕМЕСЕ "Жексенбі" болса, "Бүгін демалыс" деп шығарыңыз.
day = "Сенбі"
if day == "Сенбі" or day == "Жексенбі":
  print("Бүгін демалыс")
else:
  print("Бүгін демалыс емес")
```
